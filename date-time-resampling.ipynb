```python
import pandas as pd
# from datetime import datetime

DATETIME_FORMAT = '%Y-%m-%d %I-%p'

# d_parser = lambda x: datetime.strptime(x, DATETIME_FORMAT)
d_parser = lambda x: pd.to_datetime(x, format=DATETIME_FORMAT)

df = pd.read_csv('ETH_1h.csv', parse_dates=['Date'], date_parser=d_parser)

df.head()
```

- similar to '.str'

```python
df['DayOfWeek'] = df['Date'].dt.day_name()

df.head(3)
```

```python
minDate = df['Date'].min()
maxDate = df['Date'].max()

print(minDate)
print(maxDate)
print(maxDate - minDate)
```

### Filter By time

```python
# period_filter = (df['Date'] >= '2019') & (df['Date'] < '2020')
# same as
period_filter = (df['Date'] >= pd.to_datetime('2019-01-01')) & (df['Date'] < pd.to_datetime('2020-01-01'))

df.loc[period_filter]
```

### set the Date column as index

```python
df.set_index('Date', inplace=True)

df.head(3)
```

- Now we can filter by DateTime

```python
df.loc['2019']
```

- Data should be sorted by date for this to work

```python
df.sort_values(by='Date', ascending=True, inplace=True)

start_date = '2020-01'
end_date = '2020-02'

df.loc[start_date:end_date]
```

### **Resampling**

```python
Highs = df['High'].resample('D').max()

Highs
```

```python
Highs['2020-01-01'] == df.loc['2020-01-01']['High'].max() # the high of the day
```

```python
%matplotlib inline
```

```python
Highs.plot()
```

### **Resampling with multiple columns**

```python
df_num = df.drop(columns=['Symbol', 'DayOfWeek'])

df_num = df_num.resample('W').agg({
    'Close': 'mean',
    'High': 'max',
    'Low': 'min',
    'Volume': 'sum',
})

df_num
```

```python
df_num['High'].plot()
df_num['Close'].plot()
df_num['Low'].plot()
```

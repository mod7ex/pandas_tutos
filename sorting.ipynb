```python
import pandas as pd

df = pd.read_json('people.json')

df
```

- some modifications

```python
df[['first_name', 'last_name']] = df['name'].str.split(' ', expand=True)

df.drop(columns='name', inplace=True)

df.rename(columns={'postalZip': 'zip_code'}, inplace=True)

df
```

```python
new_records = pd.DataFrame({
    'first_name': ['Cleo'],
    'last_name': ['Vnokova']
})

df = pd.concat([df, new_records], ignore_index=True)

df
```

### Sorting

```python
df.sort_values(by='first_name', ascending=False, inplace=True)

df
```

```python
df.sort_values(by=['first_name', 'last_name'], inplace=True) # order matters

df
```

```python
df.sort_values(by=['first_name', 'last_name'], ascending=[True, False], inplace=True) # order matters

df
```

### Reset Order

```python
df.sort_index(inplace=True)

df
```

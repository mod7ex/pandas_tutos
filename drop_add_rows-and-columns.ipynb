```python
import pandas as pd

df = pd.read_json('people.json')

df
```

```python
df.rename(columns={
    'postalZip': 'zip_code'
}, inplace=True)
```

### Adding Columns

```python
df[['first_name', 'last_name']] = df['name'].str.split(' ', expand=True)

df
```

### Dropping Columns

```python
df.drop(columns=['name'], inplace=True)

df
```

### Adding records

```python
new_record = pd.DataFrame({
    'first_name': ['Mourad'],
    'last_name': ['EL CADI']
})

df = pd.concat([df, new_record], ignore_index=True)

df
```

### Dropping Rows

```python
df.drop(index=5, inplace=True)
```

```python
condition = df['email'].str.endswith('.com')

df.drop(index=df[condition].index)
```

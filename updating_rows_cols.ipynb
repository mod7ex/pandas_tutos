> Altering using Functions

```python
# df['MAIL'].apply(len)

def upperLastName (full_name: str):
    [first_name, last_name] = full_name.split(' ')
    return ' '.join([first_name, last_name.upper()])

df['FULL_NAME'] = df['FULL_NAME'].apply(upperLastName)

df
```
